```python
import pandas as pd

df = pd.read_csv('/content/gdrive/MyDrive/music_videos_v2.csv')
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

tokenized_texts = df.lyrics

cv = CountVectorizer()
dt_matrix = cv.fit_transform(tokenized_texts)
```

## Train the Model

```python
from sklearn.decomposition import LatentDirichletAllocation

N_TOPICS = 2

lda = LatentDirichletAllocation(n_components=N_TOPICS)
lda.fit(dt_matrix)
```

## Word Topic Distributions

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic {}: ".format(topic_idx)
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)

vocabulary = cv.get_feature_names()
print_top_words(lda, vocabulary, 10)
```

# Document Topic Distribution

```python
distrubutions = lda.transform(dt_matrix)
print(distrubutions)
```

```python
topic_distributions = pd.DataFrame(distrubutions, 
                      columns=['topic_{}'.format(i) for i in range(N_TOPICS)])
topic_distributions
```
